# Trying fillna with 0

Import Simple Imputer

```python
from sklearn.impute import SimpleImputer
```

First try with all 0:

```python
X = df #.drop('PageValues', axis=1) trying to fill everything with 0

imputer = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value= 0)
X_imp = imputer.fit_transform(X)
```

```python
X.columns
```

```python
df_0 = pd.DataFrame(X_imp, columns = X.columns)
```

```python
#df_0['PageValues'] = df['PageValues']
```

```python
df_0
```

Import One Hot Encoder to transform the categorical columuns:

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot = OneHotEncoder()
X_oh = one_hot.fit_transform(df_0[['VisitorType']])
```

```python
print(X_oh.todense())
one_hot.categories_
```

```python
pd.DataFrame(data=X_oh.todense(), columns=one_hot.categories_)
```

```python
df_0_oh = df_0.drop('VisitorType', axis=1)
```

```python
df_0_oh = pd.concat([df_0_oh,pd.DataFrame(data=X_oh.todense(), columns=one_hot.categories_)], axis=1)
```

```python
df_0_oh.sort_values('PageValues',ascending=False)
```

```python
df_0_oh.isnull().sum()
```

Export the clean dataset:

```python
df_0_oh.to_csv('data/df_0_oh.csv', index=False)
```
