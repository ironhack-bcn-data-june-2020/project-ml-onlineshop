##### We need to change the data types of the columns previously named to their proper type, which is numeric. 

###### We will use the apply function and pd.to_numeric for it. 

```python
wrong_dtypes = ['Administrative_Duration',
                        'Informational_Duration',
                        'ProductRelated_Duration',
                        'PageValues']
```

```python
df[wrong_dtypes] = df[wrong_dtypes].apply(pd.to_numeric, errors='coerce')
```

```python
print(df.dtypes)

'''All the columns are the correct data type besides month, let's change that one too'''
```

```python
df.Month.replace('FEBR','Feb', inplace=True)

'''We don't have information for all the months, we are missing January & April, plus we also have some nan values'''
df.Month.isnull().sum()
```
