```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/df_0_oh.csv')
```

```python
df.head(10)
```

```python
# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True],
    'max_depth': [int(x) for x in np.linspace(10, 110, num = 11)],
    'max_features': ['auto'],
    'min_samples_leaf': [3, 5, 4],
    'min_samples_split': [2, 5, 10],
    'n_estimators': [int(x) for x in np.linspace(start = 10, stop = 1000, num = 4)]
}
```

# Implementing the model

```python
X = df.drop(columns='Revenue')
y = df['Revenue']
```

```python
from imblearn.combine import SMOTETomek

oversamp_undersamp = SMOTETomek(sampling_strategy=0.6) #default sampling strategy

X_b, y_b = oversamp_undersamp.fit_sample(X, y)

X_b.shape,y_b.shape
```

```python
forest = RandomForestClassifier()
```

```python
grid_search = GridSearchCV(estimator = forest, 
                           param_grid = param_grid, 
                           cv = 3, 
                           n_jobs = -1, 
                           verbose = 2)
```

```python
grid_search.fit(X_b,y_b)
```
